### 1. What is name of car that has the highest horsepower?  

```python
#your code here
import pandas as pd
df = pd.read_csv('Cars.csv')
max_horsepower_row = df.loc[df['horsepower'].idxmax()]
car_name_highest_hp = max_horsepower_row['name']
print(f"The name of the car with the highest horsepower is: {car_name_highest_hp}")
```

### 2. How many cars have mpg  35?

```python
#your code here
high_mpg_cars = df[df['mpg'] >= 35]
count = len(high_mpg_cars)
print(f"Cars with mpg >= 35: {count}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
df = df.dropna(subset=['horsepower'])
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
common_origin = filtered_cars['origin'].mode()[0]
print(f"Most common origin: {common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
japan_cars = df[df['origin'] == 'japan']
mean_acc = japan_cars['acceleration'].mean()
mean_acc = round(mean_acc, 2)
print(f"Mean acceleration for Japan cars: {mean_acc}")
```

### 5. Which year had the highest average mpg?

```python
#your code here
avg_mpg = df.groupby('model_year')['mpg'].mean()
best_year = avg_mpg.idxmax()
print(f"Year with highest average mpg: {best_year}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
df_reset = df.reset_index()
car_counts = df_reset['name'].value_counts()
multi_year_cars = car_counts[car_counts > 1].index.tolist()
consistent_cars = []
for car_name in multi_year_cars:
    car_data = df_reset[df_reset['name'] == car_name]
    if len(car_data['model_year'].unique()) > 1:
        std_mpg = car_data['mpg'].std()
        if not pd.isna(std_mpg) and std_mpg < 1.0:
            avg_mpg = car_data['mpg'].mean()
            consistent_cars.append({'name': car_name, 'appearances': len(car_data), 'average_mpg': avg_mpg, 'std_mpg': std_mpg})
consistent_cars_df = pd.DataFrame(consistent_cars)
if not consistent_cars_df.empty:
    min_std = consistent_cars_df['std_mpg'].min()
    lowest_variation_cars = consistent_cars_df[consistent_cars_df['std_mpg'] == min_std]
    lowest_variation_cars_sorted = lowest_variation_cars.sort_values(by=['appearances', 'average_mpg'], ascending=[False, False])

    print("Car(s) with the lowest variation in mpg (std < 1.0) and multiple appearances:")
    print(lowest_variation_cars_sorted[['name', 'appearances', 'average_mpg']])
else:
    print("No consistent cars found meeting the criteria.")
```
